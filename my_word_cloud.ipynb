This script needs to process the text, remove punctuation, ignore case and words that do not contain all alphabets, count the frequencies, and ignore uninteresting or irrelevant words.  A dictionary is the output of the `calculate_frequencies` function.  The `wordcloud` module will then generate the image from your dictionary.

```python
# This is the uploader widget

def _upload():

    _upload_widget = fileupload.FileUploadWidget()

    def _cb(change):
        global file_contents
        decoded = io.StringIO(change['owner'].data.decode('utf-8'))
        filename = change['owner'].filename
        print('Uploaded `{}` ({:.2f} kB)'.format(
            filename, len(decoded.read()) / 2 **10))
        file_contents = decoded.getvalue()

    _upload_widget.observe(_cb, names='data')
    display(_upload_widget)

_upload()
```

The uploader widget saved the contents of your uploaded file into a string object named *file_contents* that your word cloud script can process.

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my","mr","mrs","into", \
"we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them","could","and", \
"their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being","for","in","on", \
"have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how","not","up","so", \
"all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just","","said"]
    
    
    # LEARNER CODE START HERE
    wordList = file_contents.split()
    wordList_WithoutPunctuations = []
    for word in wordList:
        s=""
        for letter in word:
            if letter.isalpha():
                if letter not in punctuations:
                    s+=letter
        wordList_WithoutPunctuations.append(s)
    
    for word in list(wordList_WithoutPunctuations):
        if word.lower() in uninteresting_words:
            wordList_WithoutPunctuations.remove(word)
            
    my_dict = {}
    for word in wordList_WithoutPunctuations:
        if word.upper() not in my_dict:
            my_dict[word.upper()] = 0
        my_dict[word.upper()] += 1
            
    
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(my_dict)
    return cloud.to_array()
```

```python
# Display your wordcloud image

myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
